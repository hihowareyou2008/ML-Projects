```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
titanic_data = pd.read_csv("/content/train.csv")
```

```python
titanic_data.head()
```

```python
titanic_data.shape
```

```python
titanic_data.describe()
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
```

```python
titanic_data.head()
```

```python
label_encode = LabelEncoder()
```

```python
label_1 = label_encode.fit_transform(titanic_data.Name)
label_2 = label_encode.fit_transform(titanic_data.Sex)
label_3 = label_encode.fit_transform(titanic_data.Ticket)
label_4 = label_encode.fit_transform(titanic_data.Cabin)
label_5 = label_encode.fit_transform(titanic_data.Embarked)
```

```python
titanic_data['Name'] = label_1
titanic_data['Sex'] = label_2
titanic_data['Ticket'] = label_3
titanic_data['Cabin'] = label_4
titanic_data['Embarked'] = label_5
```

```python
titanic_data.head()
```

```python
titanic_data['Survived'].value_counts()
```

```python
X = titanic_data.drop(columns='Survived', axis=1)
Y = titanic_data['Survived']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_pred = model.predict(X_train)
X_test_pred = model.predict(X_test)
training_data_accuracy = accuracy_score(X_train_pred, Y_train)
testing_data_accuracy = accuracy_score(X_test_pred, Y_test)
```

```python
print(training_data_accuracy)
print(testing_data_accuracy)
```
