```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the boston house price dataset

```python
# Fetching a built-in dataset
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset into a pandas Datframe
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
# Print first 5 rows of our dataframe
house_price_dataframe.head()
```

```python
# Add the target (price) column to the Dataframe
house_price_dataframe['Price'] = house_price_dataset.target
```

```python
# Displaying the Data
house_price_dataframe.head()
```

```python
# Checking the number of rows and columns in the dataframe
house_price_dataframe.shape
```

```python
# check for missing values
house_price_dataframe.isnull().sum()
```

```python
# statisitical measures of the dataset
house_price_dataframe.describe()
```

Understanding the coorelation between various features in the sataset
1. Positive Coorelation
2. Negative coorelation

```python
# .corr() calculates the correlation coefficent between numerical values (We are saving this correlation value within the variable "correlation")
correlation = house_price_dataframe.corr()
```

```python
# Constructing a heat map to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Splitting the data and target

```python
X = house_price_dataframe.drop(['Price'], axis=1)
Y = house_price_dataframe['Price']
```

```python
print(X)
print(Y)
```

Train, Test, Split

```python
X_Train , X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_Train.shape, X_test.shape)
```

Model Training





































XGBoost Regressor

```python
# Loading the model
model = XGBRegressor()
```

```python
# Training the model with X_Train
model.fit(X_Train, Y_train)
```

```python
# Accuracy for prediction on training data
training_data_prediction = model.predict(X_Train)
```

```python
print(training_data_prediction)
```

```python
# r squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R Squared Error: ", score_1)
print("Mean Absolute Error: ", score_2)
```

Visuallizing the actual prices and the predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual price vs predicted price")
plt.show()
```

Prediction on Test Data

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# r squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute error
score_2 = metrics.mean_absolute_error(Y_test, training_data_prediction)

print("R Squared Error: ", score_1)
print("Mean Absolute Error: ", score_2)
```
