````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
````

````python
car_data = pd.read_csv("/content/car_price_prediction_.csv")
````

````python
car_data.drop(columns='Car ID', axis=1)
````

````python
renamed_col = car_data.rename(columns={'Fuel Type':'fuel_type'}, inplace=True)
````

````python
car_data.head()
````

````python
label_encode = LabelEncoder()
````

````python
labels = label_encode.fit_transform(car_data.fuel_type)
labels_2 = label_encode.fit_transform(car_data.Mileage)
labels_3 = label_encode.fit_transform(car_data.Transmission)
labels_4 = label_encode.fit_transform(car_data.Brand)
````

````python
car_data['fuel_type'] = labels
car_data['Mileage'] = labels_2
car_data['Transmission'] = labels_3
car_data['Brand'] = labels_4
````

````python
car_data.head()
````

````python
car_data.drop(columns='Car ID', axis=1)
````

````python
car_data['Mileage'].value_counts()
````

````python
car_data['Transmission'].value_counts()
````

````python
car_data['fuel_type'].value_counts()
````

````python
X = car_data.drop(columns=['Price', 'Mileage', 'Brand'], axis=1)
Y = car_data['Price']
print(X)
print(Y)
````

````python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=2)
````

````python
print(X.shape, X_train.shape, X_test.shape)
````

````python
print(X_train)
print(Y_train)
````

````python
model = LinearRegression()
model_2 = Lasso()
````

````python
model.fit(X_train, Y_train)
model_2.fit(X_train, Y_train)
````

````python
x_train_prediction = model.predict(X_train)
x2_train_prediction = model_2.predict(X_train)
x_test_prediction = model.predict(X_test)
x2_test_prediction = model_2.predict(X_test)
````

````python
print(x_train_prediction)
print(x2_train_prediction)
````

````python
plt.scatter(Y_train, x_train_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted price")
plt.title("Actual Price Vs. Predicted Price")
plt.show()
````

````python
# r squared error
score_1 = metrics.r2_score(Y_train, x_train_prediction)
score_2 = metrics.r2_score(Y_test, x_test_prediction)
score_3 = metrics.r2_score(Y_train, x2_train_prediction)
score_4 = metrics.r2_score(Y_test, x2_test_prediction)

print(score_1)
print(score_2)
print(score_3)
print(score_4)
````
